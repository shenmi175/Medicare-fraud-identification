```python
# 加载原始数据集
file_path = "E:\竞赛\服创赛-A08\data\process_data.csv"
data = pd.read_csv(file_path)

# 查看原始数据集中出院诊断病种名称_NN列的NaN值个数
count_before_nan = data["出院诊断LENTH_MAX"].isnull().sum()

# 统计 出院诊断病种名称_NN 列值为 0 的个数
count_before_zero = data["出院诊断LENTH_MAX"].eq(0).sum()

# 将所有空值填0
data = data.fillna(0)

# 查看填充后数据集中出院诊断病种名称_NN列的NaN值个数
count_after_nan = data["出院诊断LENTH_MAX"].isnull().sum()

# 统计 出院诊断病种名称_NN 列值为 0 的个数
count_after_zero = data["出院诊断LENTH_MAX"].eq(0).sum()

# 打印统计结果
print("填充前：")
print("NaN值个数:", count_before_nan)
print("值为0的个数:", count_before_zero)
print("填充后：")
print("NaN值个数:", count_after_nan)
print("值为0的个数:", count_after_zero)
```

```python
# 计算“就诊欺诈系数”
fraud_counts_hospital = data[data['RES'] == 1].groupby('医院编码_NN').size()
total_counts_hospital = data.groupby('医院编码_NN').size()
hospital_fraud_ratio = fraud_counts_hospital / total_counts_hospital
hospital_fraud_ratio = hospital_fraud_ratio.fillna(0)
data['就诊欺诈系数'] = data['医院编码_NN'].map(hospital_fraud_ratio)

# 计算“诊断欺诈系数”
fraud_counts_diagnosis = data[data['RES'] == 1].groupby('出院诊断病种名称_NN').size()
total_counts_diagnosis = data.groupby('出院诊断病种名称_NN').size()
diagnosis_fraud_ratio = fraud_counts_diagnosis / total_counts_diagnosis
diagnosis_fraud_ratio = diagnosis_fraud_ratio.fillna(0)
data['诊断欺诈系数'] = data['出院诊断病种名称_NN'].map(diagnosis_fraud_ratio)

# 计算特征差值
data['月就诊天数差'] = data['月就诊天数_MAX'] - data['月就诊天数_AVG']
data['月就诊医院数差'] = data['月就诊医院数_MAX'] - data['月就诊医院数_AVG']
data['月统筹金额差'] = data['月统筹金额_MAX'] - data['月统筹金额_AVG']
data['月药品金额差'] = data['月药品金额_MAX'] - data['月药品金额_AVG']
data['医院_就诊天数差'] = data['医院_就诊天数_MAX'] - data['医院_就诊天数_AVG']
data['医院_统筹金差'] = data['医院_统筹金_MAX'] - data['医院_统筹金_AVG']
data['医院_药品差'] = data['医院_药品_MAX'] - data['医院_药品_AVG']

# 指定的特征列表
specified_features = ['月就诊天数差', '月就诊医院数差', '月统筹金额差', '月药品金额差', 
                      '医院_就诊天数差', '医院_统筹金差', '医院_药品差']

# 创建医院和诊断的组合特征
data['医院_诊断组合'] = data['医院编码_NN'].astype(str) + data['出院诊断病种名称_NN'].astype(str)

# 计算类别0指定特征的标准差
class_0_data = data[data['RES'] == 0]
class_0_std = class_0_data[specified_features].std()

# 调整指定特征值
for column in specified_features:
    adjusted_column_name = f'{column}_调整'
    data[adjusted_column_name] = data[column] - class_0_std[column]

# 将所有特征合并到一个数据集中
all_features = ['个人编码','就诊欺诈系数', '诊断欺诈系数', '月就诊天数差', 
               '月就诊医院数差', '月统筹金额差', '月药品金额差', '医院_就诊天数差', 
               '医院_统筹金差', '医院_药品差','医院_诊断组合'] + [f'{column}_调整' for column in specified_features]

final_selected_features = data[all_features]

# 保存到新的 CSV 文件
output_file_path = 'E:\竞赛\服创赛-A08\\data\\特征集\output_file_1.csv'  # 请替换为期望的输出文件路径
final_selected_features.to_csv(output_file_path, index=False)

print("特征保存完成。")
```

```python
import pandas as pd

data = pd.read_csv("E:\竞赛\服创赛-A08\data\特征集\process_data.csv")

# 计算类别0的平均值
class_0_data = data[data['RES'] == 0]
class_0_mean = class_0_data[['起付标准以上自负比例金额_SUM', '医疗救助个人按比例负担金额_SUM', '基本统筹基金支付金额_SUM']].mean()

# 计算类别0的四分位数以确定异常值范围
Q1 = class_0_data[['起付标准以上自负比例金额_SUM', '医疗救助个人按比例负担金额_SUM', '基本统筹基金支付金额_SUM']].quantile(0.25)
Q3 = class_0_data[['起付标准以上自负比例金额_SUM', '医疗救助个人按比例负担金额_SUM', '基本统筹基金支付金额_SUM']].quantile(0.75)
IQR = Q3 - Q1

# 定义异常值范围
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 函数来判断是否大于平均值和是否为异常值
def check_greater_than_mean_and_outlier(row, feature_name):
    is_greater_than_mean = 1 if row[feature_name] > class_0_mean[feature_name] else 2
    is_outlier = 1 if row[feature_name] < lower_bound[feature_name] or row[feature_name] > upper_bound[feature_name] else 2
    return is_greater_than_mean, is_outlier

# 应用这个函数到数据集
for feature in ['起付标准以上自负比例金额_SUM', '医疗救助个人按比例负担金额_SUM', '基本统筹基金支付金额_SUM']:
    data[f'{feature}_大于平均值'], data[f'{feature}_是异常值'] = zip(*data.apply(check_greater_than_mean_and_outlier, axis=1, feature_name=feature))


data.head()
```

```python
# 确保在所有情况下都不会产生NaN
def calculate_ratio_fixed(row, occurred_col, self_paid_col, declared_col):
    occurred_amount = row[occurred_col]
    self_paid_amount = row[self_paid_col] if self_paid_col in row and pd.notna(row[self_paid_col]) else 0
    declared_amount = row[declared_col] if declared_col in row and pd.notna(row[declared_col]) else 0

    self_paid_ratio = self_paid_amount / occurred_amount if occurred_amount != 0 else 0
    declared_ratio = declared_amount / occurred_amount if occurred_amount != 0 else 0

    return self_paid_ratio, declared_ratio

# 为每种费用计算比例并更新数据集
for expense in expense_types:
    occurred_col = f'{expense}发生金额_SUM'
    self_paid_col = f'{expense}自费金额_SUM' if f'{expense}自费金额_SUM' in data.columns else None
    declared_col = f'{expense}申报金额_SUM'

    # 为自费金额和申报金额生成新的比例列
    ratios = [calculate_ratio_fixed(row, occurred_col, self_paid_col, declared_col) for index, row in data.iterrows()]
    data[f'{expense}自费金额占比'], data[f'{expense}申报金额占比'] = zip(*ratios)

# 再次检查更新后的数据集的部分列，特别关注NaN的情况
data[['药品费自费金额占比', '药品费申报金额占比', '检查费自费金额占比', '检查费申报金额占比', '治疗费自费金额占比', '治疗费申报金额占比']].head()
```
